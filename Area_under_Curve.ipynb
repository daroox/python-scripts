# Find Area Under a Curve

```python
import numpy as np
from sklearn import metrics

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([-1, -1, 2, 2])
y = np.array([0, 4, 4, 0])
```

```python
plt.plot(x,y,':.')
plt.grid()
```

```python
fpr, tpr, thresholds = metrics.roc_curve(x, y, pos_label=2)
#metrics.auc(fpr, tpr)
```

```python
metrics.auc(x,y)
```

```python
import numpy as np
from scipy.integrate import trapz

print('{:18.16f}'.format(trapz(y, x)))
```

```python
xx = np.array([-1, -1, 2, 2, 2, 5, 5])
yy = np.array([0, 4, 4, 0,-4,-4,0])
```

```python
plt.plot(xx,yy,':.')
plt.grid()
```

```python
metrics.auc(xx,yy)
```

```python
print('{:18.16f}'.format(trapz(yy, xx)))
```

```python
x1 = np.arange(-np.pi,np.pi,0.1)
y1 = np.sin(x1)
```

```python
plt.plot(x1,y1,':.')
plt.grid();
```

```python
metrics.auc(x1,y1)
```
