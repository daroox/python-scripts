```python
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set()
#%matplotlib inline

plt.rc('text', usetex=True)
plt.rc('font', family='serif')
```

```python
    data = 'C2_25_00000.dat'
df    =   pd.read_csv(data, sep = ' ' ,names = ['Time','Ampl'])

df['time_usec'] = df['Time']*1e6
df['volt_mV']   = df['Ampl'] *-1e3 
```

```python
plt.plot(df.time_usec,df.volt_mV)
plt.xlabel(r'{Ion time of flight ($\mu$s)}')
plt.ylabel('Signal (arb. units)')
plt.yticks([])
plt.text(4, 58, r'H$_2$O$^{+}$')
plt.text(5.5, 5, r'Xe$^{2+}$')
plt.text(7.7, 33, r'Xe$^{+}$')
plt.xlim(None,15)
plt.savefig(data+'.pdf')
```
