# Baseline Subtraction and Area under the Curve

import the libraries

```python
import numpy
import peakutils
from scipy.integrate import trapz
from peakutils.plot import plot as pplot
from matplotlib import pyplot
from sklearn import metrics
%matplotlib inline
```

generat the plots and curves

```python
centers = (30.5, 72.3)
x = numpy.linspace(0, 120, 121)
y = (peakutils.gaussian(x, 5, centers[0], 3) +
peakutils.gaussian(x, 7, centers[1], 10) +
numpy.random.rand(x.size))
```

```python
y2 = y + numpy.polyval([0.002,-0.08,5], x)
pyplot.figure(figsize=(10,6))
pyplot.plot(x, y2)
pyplot.title("Data with baseline")
```

evaluate the baseline using `peakutils.baseline()`

```python
baseline_values = peakutils.baseline(y2)
```

```python
pyplot.figure(figsize=(10,6))
pyplot.plot(x, y2,x,baseline_values);
```

subtract the baseline from y-values

```python
y3 = y2 - baseline_values
```

```python
pyplot.figure(figsize=(10,6))
pyplot.plot(x, y3);
pyplot.grid()
```

Now use `sklearn` and ` trapz()` to calculate the area under the whole curve.

```python
metrics.auc(x,y3)
```

```python
trapz(y3,x)
```

 Next step is to calculate the area under one curve.
 Find out the limits of the peak

```python
pyplot.figure(figsize=(5,5))
pyplot.plot(x[20:40], y3[20:40],'r-',x,y3,'.');
pyplot.grid()
```

```python
metrics.auc()
```

calculate the area under the curve with `scipy` with syntax `trapz(y, x)`

```python
trapz(y3[20:40],x[20:40])
```
